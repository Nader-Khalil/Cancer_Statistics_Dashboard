## Features Discription:


1.   VAR: The original OECD code used for the cancer type
2.   Cancer_Type: One of the five types of cancer included or all cancer types
3. Year: The year corresponding to the value for indicator, gender and type of cancer. 
4. Country: Specifies whether is incidence or frequency.
5. COU: The original OECD code used for the indicator and gender.
6. Country_Status: Specifies if the country is member or partner or has an accession agreement with OECD
7. Measure: Specifies whether is incidence or frequency
8. UNIT: The original OECD code used for the indicator and gender
9. Value: The calculated value for the indicator taking into consideration the gender and cancer type.
